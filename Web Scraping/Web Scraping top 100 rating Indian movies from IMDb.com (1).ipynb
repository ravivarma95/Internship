```python
from bs4 import BeautifulSoup
import requests
```

```python
Page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=461131e5-5af0-4e50-bee2-223fad1e00ca&pf_rd_r=XTSKVRSARYNJG56J6GJ8&pf_rd_s=center-1&pf_rd_t=60601&pf_rd_i=india.toprated&ref_=fea_india_ss_toprated_india_tr_india250_sm')
```

```python
soup = BeautifulSoup(Page.content)
```

```python
Page
```

```python
soup
```

```python
Top_movies= soup.find_all('td', class_="titleColumn")
Top_movies = Top_movies[0:100]
Top_movies
```

```python
Top100_Movies = []
for i in Top_movies:
    Top100_Movies.append(i.find('a').text)
Top100_Movies
```

```python
Rating = soup.find_all('td', class_="ratingColumn imdbRating")
Rating = Rating[0:100]
Rating
```

```python
Top100_Rating=[] 
for i in Rating:
    Top100_Rating.append(i.text.replace('\n',''))
Top100_Rating
```

```python
Year_of_release = soup.find_all('span', class_="secondaryInfo")
Year_of_release = Year_of_release[0:100]
Year_of_release
```

```python
Top100_Years=[] 
for i in Year_of_release:
    Top100_Years.append(i.text)
Top100_Years
```

```python
top100_IMDB = {'Names':Top100_Movies,'Ratings':Top100_Rating,'Years':Top100_Years}
import pandas as pd
IMDB_Top100_Indian_movies= pd.DataFrame.from_dict(top100_IMDB, orient='index').T
IMDB_Top100_Indian_movies
```
